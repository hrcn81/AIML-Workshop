```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/hrcn81/AIML-Workshop/main/Datasets/covid_19_clean_complete_till_23_June.csv',parse_dates = ['Date'])

df.head()
```

```python
df.info()
```

```python
df.drop(['Province/State'],axis = 1,inplace = True)
```

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
```

```python
grouped_data = df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum()
```
